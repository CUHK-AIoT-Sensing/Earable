```python
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
from matplotlib import rc
rc('text', usetex=False)
plt.rcParams.update({'font.size': 20})
```

```python
OOD = 10.7
results_name = ['English', 'Self', 'Environment',]
results_metric = [8.1, 8.8, 13.2]
fig, ax = plt.subplots(figsize=(8, 6))
ax.bar(results_name, results_metric, color='steelblue', edgecolor='black', width=0.5)
# Add horizontal line for OOD
ax.axhline(y=OOD, color='red', linestyle='--', label='OOD')
ax.set_ylabel('SISNR Improvement (dB)')
plt.tight_layout()
plt.savefig('cross_noise_teacher.pdf', bbox_inches='tight')
```
